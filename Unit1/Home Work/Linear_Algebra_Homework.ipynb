**Check if the matrix is one or many of the special types we have seen (diagonal, triagonal...)**

```python
def is_diagonal(matrix):

#we first mask the diagonl(with zeros) and then we count the nonzero, if the result is 0 then our matrix is diagonal
 return np.count_nonzero(matrix - np.diag(np.diagonal(matrix))) == 0
```

```python
is_diagonal(data)
```

```python
#Test when is_diagonal is true
diagonal_matrix = np.array([[3, 0, 0],
                            [0, 5, 0],
                            [0, 0, 4]])
is_diagonal(diagonal_matrix)
```

```python
def is_upper(matrix):
  #we mask the upper part with 0, if it's upper then it's diagonal
  up_matrix = np.tril(matrix, 0)
  return is_diagonal(up_matrix)
```

```python
is_upper(data)
```

```python
up_matrix = np.array([[3, 2, 4],
                      [0, 5, 3],
                      [0, 0, 4]])
is_upper(up_matrix)
```

```python
def is_lower(matrix):
  #we mask the lower part with 0, if it's lower then it's diagonal
  lower_matrix = np.triu(matrix, 0)
  return is_diagonal(lower_matrix)
```

```python
is_lower(data)
```

```python
lower_matrix = np.array([[3, 0, 0],
                        [5, 5, 0],
                        [9, 1, 6]])
is_lower(lower_matrix)
```
